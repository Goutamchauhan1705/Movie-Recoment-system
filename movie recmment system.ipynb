```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
       L.append(i['name'])
       return L 
```

```python
movies.dropna(inplace=True) 
```

```python
movies['genres'] = mov['genres'].apply(convert)
movies.head(2)
```

```python
movies['keywords'] = mov['keywords'].apply(convert)
movies.head(2)
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python

movies['cast'] = mov['cast'].apply(convert)
movies.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
#movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.sample(5)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = mov['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = mov['genres'].apply(collapse)
movies['keywords'] = mov['keywords'].apply(collapse)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```
